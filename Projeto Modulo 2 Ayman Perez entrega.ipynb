Mas, reforçando: são 4 batidas de ponto:

- entrada (começo do dia);
- saída (sai pro almoço);
- entrada (volta do almoço);
- saída (fim do dia).

```python
## IMPORTAMOS BIBLIOTECA

import random as rd

## CLASSE HORARIO

class Horario:
    
    def __init__ (self, hour: int, minute: int, second: int):
        
        '''
        Metodo construtor para definir a classe Horario
        
        '''
        
        self.hora = hour
        self.minuto = minute
        self.segundo = second   
    
    
    def __sub__ (self, other):

        '''
        Metodo magico para calcular a tempo entre dos horarios (dois objetos da Classe Horario).

        Obtemos a contagem em segundos para depois fazer a diferença e posteriormente
        criar um objeto da Classe Horario com os parametros em h,m,s.

        '''

        h1 = self.hora
        m1 = self.minuto
        s1 = self.segundo

        h2 = other.hora
        m2 = other.minuto
        s2 = other.segundo

        #Convertemos todo em segundos

        tempo_inicio =  s1 + m1*60 + h1*3600
        tempo_final  =  s2 + m2*60 + h2*3600

        #Calculamos a diferença em segundos

        if tempo_inicio > tempo_final:

            dif_tempo_em_segundos = tempo_inicio - tempo_final

        else:

            dif_tempo_em_segundos = tempo_final - tempo_inicio

        #Tranformamos os segundos em horas,minutos,segundos

        hdif = dif_tempo_em_segundos//3600
        mdif = (dif_tempo_em_segundos-(3600*hdif))//60
        sdif = (dif_tempo_em_segundos-(3600*hdif))%60

        return Horario(hdif, mdif, sdif)
   
    
    def __add__(self, other):
        
        '''
        Metodo magico para calcular a soma entre dos horarios (dois objetos da Classe Horario).

        Obtemos a contagem em segundos para depois fazer a soma e posteriormente
        criar um objeto da Classe Horario com os parametros em h,m,s.

        '''
        
        h1 = self.hora
        m1 = self.minuto
        s1 = self.segundo

        h2 = other.hora
        m2 = other.minuto
        s2 = other.segundo
        
        #Convertemos todo em segundos
        
        seg_totais_horario1 = h1*3600 + m1*60 + s1
        seg_totais_horario2 = h2*3600 + m2*60 + s2

        #Calculamos a soma em segundos
        
        soma_em_seg = seg_totais_horario1 + seg_totais_horario2
        
        #Tranformamos os segundos em horas,minutos,segundos
        
        ht = soma_em_seg//3600
        mt = (soma_em_seg-(3600*ht))//60
        st = (soma_em_seg-(3600*ht))%60

        return Horario(ht,mt,st)      
       
    
    
    def __repr__ (self):
        
        '''
        Metodo magico para modificar el output dos objetos da classe Horario
        '''
        
        return "Output __repr__ {:02d}:{:02d}:{:02d}".format(self.hora,self.minuto,self.segundo)
    
    def __str__ (self):
        
        '''
        Metodo magico para modificar la funcion print para os objetos da classe Horario
        '''
        
        return "----------> {:02d}:{:02d}:{:02d}".format(self.hora,self.minuto,self.segundo)


## MENU 
## NOTA: não foquei na validacao de inputs. Se inserir horarios em forma de str ou qualquer input invalido dara erro. 
##       meu foco foi na criacão das classes objetos e funções

def abre_menu ():
      
    #Declaração das variaveis e objetos que vamos precisar e gerar um numero de funcioniario aleatorio.
    
    opcoes = ["c","a","v","s","k","f"]
    id_funcionario = rd.sample(population=range(0,101),k=1)
    obj_horario_c = None
    obj_horario_a = None
    obj_horario_v = None
    obj_horario_s = None
    
    print("="*50)
    print(f"Bem-vindo ao sistema bateponto, seu numero de funcionario é {id_funcionario}.\nEscolha uma opção".center(50))
    print("="*50)

    print("\nC - Chegada ao trabalho")
    print("A - Saida para almoço")
    print("V - Volta do almoço")
    print("S - Saida do trabalho")
    print("K - Calculo horario jornada de trabalho")
    print("F - Fecha o menu")
    
    
    opcao_menu = input()
    
    
    while opcao_menu.lower() not in opcoes:
        
        print("opçao errada, tente de novo\n")
    
        print("="*50)
        print("Escolha uma opção".center(50))
        print("="*50)

        print("\nC - Chegada ao trabalho")
        print("A - Saida para almoço")
        print("V - Volta do almoço")
        print("S - Saida do trabalho")
        print("K - Calculo horario jornada de trabalho")
        print("F - Fecha o menu")
    
        opcao_menu = input()
    
    #vamos criar 4 objetos da classe Horario para registrar os horarios de saida e entrada
    
    while opcao_menu.lower() != "f":
    
        if opcao_menu.lower() == "c":

            hc = int(input("Hora de chegada ao trabalho?"))
            mc = int(input("minuto de chegada ao trabalho?"))
            sc = int(input("segundo de chegada ao trabalho?"))
            
            obj_horario_c = Horario(hc,mc,sc)
            
            
        elif opcao_menu.lower() == "a":

            ha = int(input("Hora de saida para almoço?"))
            ma = int(input("minuto de saida para almoço?"))
            sa = int(input("segundo de saida para almoço?"))
            
            obj_horario_a = Horario(ha,ma,sa)
            
            
        elif opcao_menu.lower() == "v":

            hv = int(input("Hora de chegada do almoço?"))
            mv = int(input("minuto de chegada do almoço?"))
            sv = int(input("segundo de chegada do almoço?"))
            
            obj_horario_v = Horario(hv,mv,sv)
            
            #impedimos registrar volta do almoço menor a uma hora usando o atributo -hora- do nosso objeto da Classe Horario
            
            demora_almoco = obj_horario_v - obj_horario_a
            if demora_almoco.hora < 1:
            
                print("="*50)
                print("ATENÇÃO, horario invalido, tempo minimo para almoçar é de 1 hora\nTente de novo")
                hv = int(input("Hora de chegada do almoço?"))
                mv = int(input("minuto de chegada do almoço?"))
                sv = int(input("segundo de chegada do almoço?"))
            
                obj_horario_v = Horario(hv,mv,sv)
                

        elif opcao_menu.lower() == "s":

            hs = int(input("Hora de saida?"))
            ms = int(input("minuto de saida?"))
            ss = int(input("segundo de saida?"))
            
            obj_horario_s = Horario(hs,ms,ss)
            
            
        elif opcao_menu.lower() == "k":
            
            #checamos primeiro se os 4 objetos da classe Horario ja foram preenchidos
            
            if obj_horario_c == None or obj_horario_a == None or obj_horario_v == None or obj_horario_s == None:
                
                print("="*50)
                print("ATENÇAO!!!\n Preencha todos os horarios entes de calcular seu dia de trabalho")
                print("="*50)
                print(f"Horario Chegada = {obj_horario_c}")
                print(f"Horario saida Almoço = {obj_horario_a}")
                print(f"Horario volta Almoço = {obj_horario_v}")
                print(f"Horario saida do Trabalho = {obj_horario_s}")
                
                     
            else:

                #calculamos o tempo trabalhado total usando os metodos magicos __add__ e __sub__ 

                obj_total_manha = obj_horario_c - obj_horario_a
                obj_total_tarde = obj_horario_v - obj_horario_s
                obj_total = (obj_horario_v - obj_horario_s) + (obj_horario_c - obj_horario_a)

                print(f"Voce chegou as {obj_horario_c} ao trabalho")
                print(f"Voce demorou {obj_horario_a - obj_horario_v} para almoçar")
                print(f"Voce trabalhou {obj_total_manha} de manha")
                print(f"mais {obj_total_tarde} de tarde")
                print(f"TOTAL de {obj_total} hor/min/seg")
                print(f"saiu do trabalho as {obj_horario_s}")      
                
                #verificamos se o funcionario trabalhou mais de 10 horas usando o atributo -hora- do nosso objeto da Classe Horario
                
                if obj_total.hora >=10:
                    print("="*50)
                    print("\nATENCÃO, voce trabalhou mais de 10 horas!!! nao faça mais isso ou será punido")
                
                return "Menu fechado, ate a proxima"       
        
        print("="*50)
        print(f"Continuamos seu registro de horarios.\nEscolha uma opção".center(50))
        print("="*50)

        print("\nC - Chegada ao trabalho")
        print("A - Saida para almoço")
        print("V - Volta do almoço")
        print("S - Saida do trabalho")
        print("K - Calculo horario jornada de trabalho")
        print("F - Fecha o menu")
    
    
        opcao_menu = input("")
        
    
    return "Menu fechado, ate a proxima"
    

    
abre_menu()



    
```
